```python
from deepeval import evaluate
from deepeval.metrics import GEval, AnswerRelevancyMetric, FaithfulnessMetric
from deepeval.test_case import LLMTestCase, LLMTestCaseParams

def evaluate_correctness(question, response):
  metric = GEval(
    name="Correctness",
    criteria="Determine whether the actual output is factually correct based on the expected output.",
    evaluation_steps=[
        "Check whether the facts in 'actual output' contradicts any facts in 'expected output'",
        "You should also heavily penalize omission of detail",
        "Vague language, or contradicting OPINIONS, are OK"
    ],
    evaluation_params=[LLMTestCaseParams.ACTUAL_OUTPUT],
  )

  test_case = LLMTestCase(
    input=question,
    actual_output=response,
  )

  metric.measure(test_case)
  print("Correctness score: ", metric.score)
  print("Reason: ", metric.reason)


"""
The answer relevancy metric uses LLM-as-a-judge to measure the quality of your RAG pipeline's generator
by evaluating how relevant the actual_output of your LLM application is compared to the provided input.
Deepeval's answer relevancy metric is a self-explaining LLM-Eval, meaning it outputs a reason for its metric score.
"""
def evaluate_relevancy(question, response):
  metric = AnswerRelevancyMetric(
    threshold=0.7,
    model="gpt-4o",
    include_reason=True
  )
  test_case = LLMTestCase(
    input=question,
    actual_output=response
  )

  metric.measure(test_case)
  print("Relevancy score: ", metric.score)
  print("Reason: ", metric.reason)


"""
The faithfulness metric uses LLM-as-a-judge to measure the quality of your RAG pipeline's generator
by evaluating whether the actual_output factually aligns with the contents of your retrieval_context.
deepeval's faithfulness metric is a self-explaining LLM-Eval, meaning it outputs a reason for its metric score.
"""
def evaluate_faithfullness(question, response, context):
  metric = FaithfulnessMetric(
    threshold=0.7,
    model="gpt-4o",
    include_reason=True
  )
  test_case = LLMTestCase(
    input=question,
    actual_output=response,
    retrieval_context=context
  )

  metric.measure(test_case)
  print("Faithfullness score: ", metric.score)
  print("Reason: ", metric.reason)


# evaluate(test_cases=[test_case], metrics=[metric])
```

# Base model

```python
def generate_output(query):
  print(f"Question: {query}")

  messages = [
      {"role": "user", "content": query},
  ]
  input_ids = tokenizer.apply_chat_template(
      messages,
      add_generation_prompt=True,
      return_tensors="pt"
  ).to(model.device)

  terminators = [
      tokenizer.eos_token_id,
      tokenizer.convert_tokens_to_ids("<|eot_id|>")
  ]

  outputs = model.generate(
      input_ids,
      max_new_tokens=256,
      eos_token_id=terminators,
      do_sample=False,
      # temperature=0.6, # If do_sample=True
      # top_p=0.9,  # If do_sample=True
  )

  response = outputs[0][input_ids.shape[-1]:]
  decoded_response = tokenizer.decode(response, skip_special_tokens=True)
  print(f"Answer: \n{decoded_response}\n")

  print("Evaluation:")
  evaluate_correctness(query, decoded_response)
  evaluate_relevancy(query, decoded_response)
  print("--------------------------------------------------------------")

  return decoded_response
```

# RAG Implementation

```python
from sentence_transformers import SentenceTransformer

emb_model = SentenceTransformer("infly/inf-retriever-v1-1.5b", trust_remote_code=True)
# In case you want to reduce the maximum length:
emb_model.max_seq_length = 4096
```

```python
documents = []
rag_data_name_list = ["abenomics.txt", "caffeine.txt", "gold_card.txt", "llm.txt", "ozempic.txt", "vibe_coding.txt"]
for filename in rag_data_name_list:
  with open(f"/content/drive/MyDrive/ai_engineering_2025/lecture-ai-engineering/day3/data/{filename}", "r") as f:
    raw_writedown = f.read()
    documents += [text.strip() for text in raw_writedown.split("。")]

print("ドキュメントサイズ: ", len(documents))
```

```python
def generate_rag_output(query):
  messages = [
      {"role": "user", "content": query},
  ]
  input_ids = tokenizer.apply_chat_template(
      messages,
      add_generation_prompt=True,
      return_tensors="pt"
  ).to(model.device)

  terminators = [
      tokenizer.eos_token_id,
      tokenizer.convert_tokens_to_ids("<|eot_id|>")
  ]

  outputs = model.generate(
      input_ids,
      max_new_tokens=256,
      eos_token_id=terminators,
      do_sample=False,
      # temperature=0.6, # If do_sample=True
      # top_p=0.9,  # If do_sample=True
  )

  response = outputs[0][input_ids.shape[-1]:]
  decoded_response = tokenizer.decode(response, skip_special_tokens=True)

  return decoded_response
```

```python
def generate_with_topk_references(question, topk):
  query_embeddings = emb_model.encode([question], prompt_name="query")
  document_embeddings = emb_model.encode(documents)
  scores = (query_embeddings @ document_embeddings.T) * 100

  print(f"Question: {question}")
  print("--------------------------------------------------------------")

  context = []
  for i, index in enumerate(scores.argsort()[0][::-1][:topk]):
    context.append(documents[index])
    print(f"取得したドキュメント{i+1}: (Score: {scores[0][index]})")
    print(documents[index], "\n\n")
  print("--------------------------------------------------------------")

  print("Answer:")
  references = "\n".join(["* " + documents[i] for i in scores.argsort()[0][::-1][:topk]])
  query =  f"[参考資料]\n{references}\n\n[質問] {question}"
  response = generate_rag_output(query)
  print(response)
  print("--------------------------------------------------------------")
  print("Evaluation:")
  evaluate_correctness(question, response)
  evaluate_relevancy(question, response)
  evaluate_faithfullness(question, response, context)
```
